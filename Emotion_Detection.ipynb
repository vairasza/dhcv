```python
import torch, torchvision

# import some common libraries
import numpy as np
import os, json, cv2, random
from google.colab.patches import cv2_imshow
import pandas as pd

from fer import FER
```

```python
detectorMT = FER(mtcnn=True)
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
emotions = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']
categories = ['box', 'emotions']
```

```python
frame = []
person = []
box = []
angry = []
disgust = []
fear = []
happy = []
sad = []
suprise = []
neutral = []
```

```python
directory = os.fsencode("/content/gdrive/MyDrive/digital_humanities/Walkthroughts/Siedler_Frames/")

for file in os.listdir(directory):
  filename = os.fsdecode(file)
  img = cv2.imread("/content/gdrive/MyDrive/digital_humanities/Walkthroughts/Siedler_Frames/" + filename)
  prediction = detectorMT.detect_emotions(img)
  
  for n in range(len(prediction)):
    bbox = prediction[n][categories[0]]
    img = cv2.rectangle(img, (bbox[0], bbox[1]), (bbox[0] + bbox[2], bbox[1] + bbox[3]),(0, 255, 0), 1,)
    cv2.putText(img, str(n), (bbox[0] + 2, bbox[1] + 15), cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,255,0),1,cv2.LINE_AA, )
    frame.append(filename)
    person.append(n)
    box.append(bbox)
    angry.append(prediction[n][categories[1]][emotions[0]])
    disgust.append(prediction[n][categories[1]][emotions[1]])
    fear.append(prediction[n][categories[1]][emotions[2]])
    happy.append(prediction[n][categories[1]][emotions[3]])
    sad.append(prediction[n][categories[1]][emotions[4]])
    suprise.append(prediction[n][categories[1]][emotions[5]])
    neutral.append(prediction[n][categories[1]][emotions[6]])

  if len(prediction) > 0:
    cv2.imwrite("/content/gdrive/MyDrive/digital_humanities/Ergebnisse/Emotion/Siedler/" + filename, img)
  
```

```python
df = pd.DataFrame(list(zip(frame, person, box, angry, disgust, fear, happy, sad, suprise, neutral)))
df.columns = ['Name', 'person', 'box', 'angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']
df.to_csv("/content/gdrive/MyDrive/digital_humanities/Ergebnisse/Emotion/siedler_emotion.csv")
```
