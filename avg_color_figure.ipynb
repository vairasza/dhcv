```python
# get rgb color values from the row data for the game
def getAvgColorValuesArr(row_data_path):
    color_values = row_data_path['RGB']
    color_values_arr = []
 
    for i in color_values:
        color_value = tuple(float(num) for num in i.replace('(', '').replace(')', '').replace('...', '').split(', '))
        color_values_arr.append(color_value)
    
    avg_color_values_arr = tuple(tuple(map(int, tup)) for tup in color_values_arr)
    
    return avg_color_values_arr
```

```python
settler_values = getAvgColorValuesArr(avg_color_data_settler_3)
diablo_values = getAvgColorValuesArr(avg_color_data_diablo_2)
elden_ring_values = getAvgColorValuesArr(avg_color_data_elden_ring)
half_life_values = getAvgColorValuesArr(avg_color_data_half_life_2)
little_nightmares_values = getAvgColorValuesArr(avg_color_data_little_nightmares)
nfsmw_values = getAvgColorValuesArr(avg_color_data_nfsmw)
starcraft_values = getAvgColorValuesArr(avg_color_data_starcraft_2)
mario_values = getAvgColorValuesArr(avg_color_data_super_mario_world)
```
